```python
df = pd.read_excel('sample_pivot.xlsx')
```

```python
pd.pivot_table(df,index = 'Region',values = 'Units')
```

```python
df['Type'].value_counts()
```

```python
get_second_highest = lambda x: pd.DataFrame(x).sort_values(by = 'Sales',ascending=False).iloc[1]

# Apply the lambda function to the 'category' column within each group
result = df.groupby('Type')['Region'].agg(get_second_highest).reset_index()
result
```
