```python
import pandas as pd
import numpy as np
```

# Adding periods parameter
adding sample line

```python
np.random.seed(20)
df = pd.DataFrame(np.random.randn(10,4), 
             index = pd.date_range('2021-07-18',periods = 10),
             columns  = ['a','b','c','d'])
```

```python
df
```

### adding new column 

```python
df['new'] = df['c']+df['d']
df
```

## Apply function 

```python
df['new_col_sqaure_d'] = df['d'].apply(np.square) 
```

```python
df
```

```python
df['d_log'] = df['d'].apply(lambda x : np.log(x))
df
```

```python
def get_logs(x):
    if (x<0):
        return x 
    else:
        return np.log(x)
df['a_log'] = df['a'].apply(get_logs)
```

```python
df
```

```python
def get_signs(x):
    if (x<0):
        return 0
    else:
        return 1
df['signs_a'] =df['a'].apply(get_signs)
```
