# Regular expressions

### Fundamental Notes 
 * '\n' denotes a newline 
 * '\t' denotes a tab
 * 'r' expression that voids the python's special characters  

```python
import re 
re.search('n','\n') # first item is pattern and second is the string
```

```python
re.search('n',r'\n') # adding backslash  -- raw string 
```

## Regular expression special character

```python
re.search(r'\n','\n\n') # '\n' is always a newline in regex (first part)
```

```python
re.search('\n','\n\n') # match found 
```

```python
re.search('\n',r'\n\n')
```

## Printing outputs of match and search

```python
re.search('n.+','abcdefnc ejfs').group()
```

```python
re.search('c','thomascdfse').start()
```

## Character sets

 * \w matches alpha numeric characters [a-zA-Z0-9_]
 * \W is the compliment of \w

```python
re.search('\w\w\w\w','ab_cdef thomas')
```

```python
re.search('\w\w\w','a3.!-!')
```

```python
re.search(r'\w\w\w','a33-_!').group()
```

```python
re.search('\W\W\W','a3. -_!') # empty space is also a character 
```

## Quantifiers 

 * '+' = 1 or more 
 * '?" = 0 or 1 
 * '*' = 0 or more 
 * [n,m] = n to m repetitions (,3),(3,) 



```python
re.search('\w+','hello world') # greedy quantifier 
```

```python
re.search('\w+\W+\w+','thomaskutty   reji')
```

```python
re.search('\w+\W?\w+','thomas kutty reji')
```

```python
re.search('\w{3}','thomaskutty reji')
```
